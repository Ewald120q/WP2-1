```python
import matplotlib.pyplot as plt
import numpy as np

# 1. Die Daten Schicht für Schicht
layers = [
    'Conv2D_1\n(8, 5x5)', 
    'Conv2D_2\n(8, 5x5)', 
    'Conv2D_3\n(12, 5x5)', 
    'Dense_1\n(256)', 
    'Dense_2\n(2)'
]
params = [208, 1608, 2412, 9981184, 514]

# 2. Plot initialisieren
plt.figure(figsize=(12, 7))

# Balken zeichnen (in deinem schönen Blau)
bars = plt.bar(layers, params, color='#009ee3', edgecolor='black', linewidth=1.2, zorder=3)

# 3. Skalierung anpassen (logarithmisch, da die Werte extrem weit auseinander liegen)
plt.yscale('log')

# 4. Exakte Zahlen über die Balken schreiben (Der saubere Weg!)
formatiert = [f'{val:,}'.replace(',', '.') for val in params]
plt.bar_label(bars, labels=formatiert, padding=5, fontsize=11, fontweight='bold')

# 5. Achsenbeschriftung und Titel
plt.title('Parameter-Verteilung pro Schicht', fontsize=16, pad=20)
plt.ylabel('Anzahl Parameter (Logarithmische Skala)', fontsize=14)
plt.xlabel('Netzwerk-Schichten', fontsize=14)

# x-Achsen-Labels leicht drehen, falls nötig
plt.xticks(fontsize=11)
plt.yticks(fontsize=11)

# Ein feines Raster im Hintergrund
plt.grid(axis='y', linestyle='--', alpha=0.6, which='major', zorder=0)

# 6. Plot anzeigen und optional speichern
plt.tight_layout()

# Wenn du das Bild direkt für LaTeX/Word speichern willst:
# plt.savefig('parameter_verteilung.png', dpi=300, bbox_inches='tight')

plt.show()
```
