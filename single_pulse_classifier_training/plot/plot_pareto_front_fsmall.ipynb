```python
import numpy as np
from scipy import stats
import statistics

# Deine Daten
params = [4254, 4838, 5422, 6730, 11554, 13874, 20290]
runs = [
    [0.7394, 0.7419, 0.7301], # 3_GAP
    [0.7263, 0.6514, 0.7329], # 4_GAP
    [0.7239, 0.7305, 0.7373], # 5_GAP
    [0.7295, 0.7327, 0.6732], # 6_GAP
    [0.7325, 0.7215, 0.7263], # 7_GAP
    [0.7446, 0.6370, 0.7395], # 8_GAP
    [0.7367, 0.6348, 0.7186]  # 9_GAP
]

# Mediane berechnen
medians = [statistics.median(r) for r in runs]

# Spearman Rangkorrelation berechnen
# rho: Korrelationskoeffizient (-1 bis 1)
# p_value: Statistische Signifikanz
rho, p_value = stats.spearmanr(params, medians)

print(f"Mediane: {medians}")
print(f"Spearman-Rho: {rho:.4f}")
print(f"p-Wert: {p_value:.4f}")
```

```python
import matplotlib.pyplot as plt

# 1. Deine Daten (Name, Parameteranzahl, Accuracy)
#data = [
#    ("3_GAP_dropout", 4254, 0.7366),
#    ("4_GAP_dropout", 4838, 0.7425),
#    ("5_GAP_dropout", 5422, 0.7335),
#    ("6_GAP_dropout", 6730, 0.7439),
#    ("7_GAP_dropout", 11554, 0.7446),
#    ("8_GAP_dropout", 13874, 0.7376),
#    ("9_GAP_dropout", 20290, 0.7376)
#]  #dropout

data = [
    ("3_GAP", 4254, 0.7399), #1e-4
    ("4_GAP", 4838, 0.7451), #1e-4
    ("5_GAP", 5422, 0.7337), #1e-5
    ("6_GAP", 6730, 0.7393), #1e-4
    ("7_GAP", 11554, 0.7328), #1e-4
    ("8_GAP", 13874, 0.7336), #1e-4
    ("9_GAP", 20290, 0.7426), #1e-4
    
    #("3_GAP_dropout", 4254, 0.7366),
    #("4_GAP_dropout", 4838, 0.7425),
    #("5_GAP_dropout", 5422, 0.7335),
    #("6_GAP_dropout", 6730, 0.7439),
    #("7_GAP_dropout", 11554, 0.7446),
    #("8_GAP_dropout", 13874, 0.7376),
    #("9_GAP_dropout", 20290, 0.7376)
]

# 2. Daten sortieren: Zuerst nach Parameteranzahl (aufsteigend)
# Für die Pareto-Front wollen wir Parameter MINIMIEREN und Accuracy MAXIMIEREN
data.sort(key=lambda x: x[1])

pareto_front = []
max_acc_so_far = -1.0

# 3. Pareto-Front berechnen
# Ein Punkt ist auf der Pareto-Front, wenn er eine höhere Accuracy hat 
# als alle Modelle, die WENIGER oder GLEICH VIELE Parameter haben.
for name, params, acc in data:
    if acc > max_acc_so_far:
        pareto_front.append((name, params, acc))
        max_acc_so_far = acc

# Listen für den Plot entpacken
all_params = [x[1] for x in data]
all_accs = [x[2] for x in data]

pareto_params = [x[1] for x in pareto_front]
pareto_accs = [x[2] for x in pareto_front]

# 4. Plotten
plt.figure(figsize=(10, 6))

# Alle Punkte als graue Punkte plotten (Hintergrund)
plt.scatter(all_params, all_accs, color='lightgray', label='Alle Modelle', s=50)

# Pareto-Punkte hervorheben
plt.scatter(pareto_params, pareto_accs, color='blue', label='Pareto-Front', s=80, zorder=5)

# Die Pareto-Kurve zeichnen (verbindet die besten Modelle)
plt.plot(pareto_params, pareto_accs, color='blue', linestyle='--', zorder=4)

# Labels zu den einzelnen Punkten hinzufügen
for name, params, acc in data:
    # Ausreißer etwas blasser labeln
    color = 'black' if acc >= 0.7 else 'gray'
    plt.annotate(name, (params, acc), textcoords="offset points", xytext=(0,10), ha='center', color=color)

# Achsenbeschriftung und Titel (Schön formatieren für die Thesis!)
plt.title('Trade-off: Parameteranzahl vs. Validation Accuracy (f_small)')
plt.xlabel('Anzahl Parameter')
plt.ylabel('Validation Accuracy')
plt.grid(True, linestyle=':', alpha=0.7)
plt.legend()

# Plot anzeigen
plt.tight_layout()
plt.show()
```

# f_small weight decay

```python
import os
import matplotlib.pyplot as plt
from tensorboard.backend.event_processing.event_accumulator import EventAccumulator

def read_tensorboard_scalar(log_dir, tag="Accuracy"):
    """Liest die X- und Y-Werte einer Kurve aus."""
    event_acc = EventAccumulator(log_dir)
    event_acc.Reload()
    
    if tag in event_acc.Tags().get('scalars', []):
        events = event_acc.Scalars(tag)
        return [e.step for e in events], [e.value for e in events]
    return [], []

def plot_all_in_one(wd_runs, model_name="3_GAP"):
    # Ein einziger, großer Plot
    plt.figure(figsize=(12, 7))
    
    # 4 deutliche, farbenblinden-freundliche Farben für deine 4 Runs
    colors = ["#1f77b4", "#ff7f0e", "#2ca02c", "#d62728"] # Blau, Orange, Grün, Rot
    
    for idx, (wd_label, wd_base_path) in enumerate(wd_runs.items()):
        color = colors[idx % len(colors)]
        
        # Pfade zu Train und Val
        path_train = os.path.join(wd_base_path, "Accuracy_train")
        path_val = os.path.join(wd_base_path, "Accuracy_val")
        
        steps_train, vals_train = read_tensorboard_scalar(path_train, tag="Accuracy")
        steps_val, vals_val = read_tensorboard_scalar(path_val, tag="Accuracy")
        
        # 1. Training plotten (Gestrichelt, leicht transparent)
        if steps_train and vals_train:
            plt.plot(steps_train, vals_train, color=color, linestyle="--", 
                     alpha=0.6, linewidth=1.5, label=f"Train (WD={wd_label})")
            
        # 2. Validierung plotten (Durchgezogen, fett)
        if steps_val and vals_val:
            plt.plot(steps_val, vals_val, color=color, linestyle="-", 
                     alpha=0.9, linewidth=2.5, label=f"Val (WD={wd_label})")

    # --- Paper-Ready Styling ---
    plt.title(f"Grid Search: Weight Decay für Modell {model_name}", fontsize=16, pad=15)
    plt.xlabel("Epoche", fontsize=14)
    plt.ylabel("Accuracy", fontsize=14)
    
    plt.grid(True, linestyle=":", alpha=0.7, color="gray")
    
    # Legende mit 2 Spalten (links Train, rechts Val) für bessere Übersicht
    plt.legend(fontsize=11, loc="lower right", ncol=2, framealpha=0.9)
    
    plt.tight_layout()
    
    # Als Vektorgrafik speichern
    dateiname = f"WD_SinglePlot_{model_name}.pdf"
    plt.savefig(dateiname, format="pdf", bbox_inches="tight")
    print(f"\n✅ Plot mit 8 Linien erfolgreich als '{dateiname}' gespeichert!")
    
    plt.show()

# ==========================================
# HIER DEINE 4 PFADE EINTRAGEN
# ==========================================
meine_wd_runs = {
    "0.0": "/cephfs/users/oleksjuk/MA/WP2-1/single_pulse_classifier_training/tensorboard_runs_grid_08_01_01_new/DM_time_binary_classificator_241002_3_GAP_dropout0.0-dmt_100_15_b128_wd0.0/lr0.0001_wd0.0",
    "1e-3": "/cephfs/users/oleksjuk/MA/WP2-1/single_pulse_classifier_training/tensorboard_runs_grid_08_01_01_new/DM_time_binary_classificator_241002_3_GAP_dropout0.0-dmt_100_15_b128_wd0.001/lr0.0001_wd0.001",
    "1e-4": "/cephfs/users/oleksjuk/MA/WP2-1/single_pulse_classifier_training/tensorboard_runs_grid_08_01_01_new/DM_time_binary_classificator_241002_3_GAP_-dmt_100_15/lr0.0001_wd0.0001",
    "1e-5": "/cephfs/users/oleksjuk/MA/WP2-1/single_pulse_classifier_training/tensorboard_runs_grid_08_01_01_new/DM_time_binary_classificator_241002_3_GAP_dropout0.0-dmt_100_15_b128_wd1e-05/lr0.0001_wd1e-05"
}





plot_all_in_one(meine_wd_runs, model_name="3_GAP")
```
